```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
```

```python
X,y=datasets.make_regression(n_samples=100,n_features=1)
X=torch.from_numpy(X.astype(np.float32)).reshape(X.shape[0],-1)
y=torch.from_numpy(y.astype(np.float32)).reshape(y.shape[0],-1)
```

```python
class LinearRegression(nn.Module):
    def __init__(self,input_dim,output_dim):
        super(LinearRegression,self).__init__()
        self.lin=nn.Linear(input_dim,output_dim)
    def forward(self,x):
          return self.lin(x)

lr=0.1
n_iter=40

model=LinearRegression(1,1)
loss=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=lr)

for epoch in range(n_iter):
  y_pred=model(X)
  l=loss(y,y_pred)
  l.backward()
  optimizer.step()
  optimizer.zero_grad()
  print(f'Epoch:{epoch+1} Loss:{l:.3f}')


  plt.figure(figsize=(5,5))
  # plt.subplot(1,1,epoch+1)
  # plt.subplot(1,40,epoch+1)
  plt.plot(X,y,'r')
  plt.plot(X,y_pred.detach(),'b')
  plt.show()

```
